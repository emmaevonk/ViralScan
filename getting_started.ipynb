# Getting Started with ViralScan
ViralScan is a command-line tool which detects viral load in paired-end FASTQ samples by using kb-python and snakemake.
In this notebook, you'll learn how to install ViralScan and how to use (and understand) this computational framework.

```python
# Installation
python -m pip install --index-url https://test.pypi.org/simple/ --no-deps viralscan

# pyfiglet needs to be installed separately
pip install pyfiglet
```

To run ViralScan succesfully you'll need a conda environment with both snakemake and kb-python.

```python
# Setting up the conda env
conda create -n conda_env -c conda-forge -c bioconda snakemake kb-python
```

```python
# Activate the conda env created
conda activate conda_env
```
